```python
X = data[[
 'acc_rf_x',
 'acc_rf_y',
 'acc_rf_z',
 'acc_rs_x',
 'acc_rs_y',
 'acc_rs_z',
 'gyro_rs_x',
 'gyro_rs_y',
 'gyro_rs_z',
 'EMG_r']]

min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X.values)
DF = pd.DataFrame(x_scaled, columns=list(X))

DF.describe()
```

```python
rbm1 = RBM(10, 18, ['rbmw1', 'rbvb1', 'rbmhb1'], 0.001)
rbm2 = RBM(18, 12, ['rbmw2', 'rbvb2', 'rbmhb2'], 0.001)
rbm3 = RBM(12, 6, ['rbmw3', 'rbvb3', 'rbmhb3'], 0.001)
rbm4 = RBM(6, 3, ['rbmw4', 'rbvb4', 'rbmhb4'], 0.001)
```

```python
def stacked_transform(stack, machine_index, batch):
    curr_batch = batch
    for rbm in stack[:machine_index]:
        curr_batch = rbm.transform(curr_batch)
    return curr_batch

def train_rbm_stack(stack, data, epochs=100, batch_size=10, file_prefix=''):
    batch_counter = 0
    batch = data[batch_counter*batch_size:(batch_counter + 1)*batch_size]
    for i in range(len(stack)):
        rbm = stack[i]
        for j in range(epochs):
            train_batch = stacked_transform(stack, i, batch)
            rbm.partial_fit(train_batch)
        rbm.save_weights('./weights/rbms/{}rbm{}.chp'.format(file_prefix, i))
                
```

```python
rbm_stack = [rbm1, rbm2, rbm3, rbm4]

train_rbm_stack(rbm_stack, X_train)
```

```python
autoencoder = AutoEncoder(10, [18, 12, 6, 3], [
    ['rbmw1', 'rbmhb1'],
    ['rbmw2', 'rbmhb2'],
    ['rbmw3', 'rbmhb3'],
    ['rbmw4', 'rbmhb4'],
], tied_weights=False)
```

```python
def load_stack_to_ae(stack, autoencoder, file_prefix=''):
    for i in range(len(stack)):
        autoencoder.load_rbm_weights(
            './weights/rbms/{}rbm{}.chp'.format(file_prefix, i),
            ['rbmw{}'.format(i + 1), 'rbmhb{}'.format(i + 1)],
            i
        )
```

```python
load_stack_to_ae(rbm_stack, autoencoder)
```

```python
def train_auto_encoder(ae, data, epochs = 500, batch_size=10):
    for i in range(epochs):
        batch = data[i*batch_size:(i + 1)*batch_size]
        ae.partial_fit(batch)
```

```python
ae_train_X = X_train[4000:]

train_auto_encoder(autoencoder, ae_train_X)
```
