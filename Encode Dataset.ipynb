```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from keras.layers import Input, Dense
from keras.models import Model, load_model
```

```python
data = pd.read_csv('data/necessary_data.csv')

data.describe()
```

```python
X = data[[
 'acc_rs_x',
 'acc_rs_y',
 'acc_rs_z',
 'gyro_rs_x',
 'gyro_rs_y',
 'gyro_rs_z',
 'EMG_r']]
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X.values)
DF = pd.DataFrame(x_scaled, columns=list(X))

DF.describe()
```

```python
DF.head()
```

```python
from keras import backend as K
import keras
```

```python
def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1)) 

keras.losses.root_mean_squared_error = root_mean_squared_error
    
autoencoder = load_model('models/one_layer_ae_rmse.h5')
autoencoder.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
encoder_input = Input(shape=(7,))
encoder_layer = autoencoder.layers[1]

encoder = Model(encoder_input, encoder_layer(encoder_input))
```

```python
decoder = load_model('models/one_layer_d_rmse.h5')
decoder.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
X = DF
optimal_features = encoder.predict(X)
```

```python
autoencoder.evaluate(X, X, batch_size=256)
```

```python
decoder.evaluate(optimal_features, X, batch_size=256)
```

```python
data['r_of_1'] = optimal_features.T[0]
data['r_of_2'] = optimal_features.T[1]
data['r_of_3'] = optimal_features.T[2]
```

```python
data['rf_ang_vel'] = data['gyro_rf_y'] - data['gyro_rs_y']
data['lf_ang_vel'] = data['gyro_lf_y'] - data['gyro_ls_y']
```

```python
data.to_csv('data/encoded_data.csv')
```
